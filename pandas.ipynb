# GroupBy Methods
Good when using categorical types 

```python
df.groupby('Airline')[['DepDelay']].mean()
```

```python
df_agg = df.groupby('Airline')[['DepDelay', 'ArrDelay']].agg(['mean','min','max'])
```

```python
df_agg.columns = ['_'.join(c) for c in df_agg.columns]
#this is list comprehension 
```
